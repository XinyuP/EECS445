(c) L1 penalty produces much sparser theta vectors for all settings of C. Since the gradient of the L1 norm is constant with respect to each element, there is just as much value in reducing a small value as there is a large value. This ends up
pushing the coefficients of the unimportant features all the way to zero thus resulting in a sparser graph. <br /> For the L2 norm, the gradient is linear with respect to each element meaning that it more value is gained from reducing large elements and less is gained from reducing small elements. This incentivizes balancing the weights of all features in order to not have expensive large values.<br /><br />

(d)Squared hinge loss is more open on the cases where a data point is classified correctly but within the margin compared to hinge loss, because the square of the number less than 1 is smaller than the number itself. On the other hand, squared hinge loss punishes misclassifications more harshly, because the square of a number larger than 1 is larger than the number itslef. So, we see a comparatively wider margin and more support vectors when using squared hinge loss.

First, I process the data using stemming (words are reduced to a root by removing inflection through dropping unnecessary characters, usually a suffix). <br />
I also used stop words and position tagging in the process of stemming.<br />
I used NLTK library for the task of stemming. <br />
I modified extract_word and extract_dictionary() so that the word_dict now contains less words which are all base form and root word. <br /><br />
Second, I use a different featuring method. The bag-of-word model is too simple. If the new comment conatins new words, the dictionary needs to grow its size, which will increase the size of the feature matrix. Also, the feature matrix may contain lots of zeros which makes the matrix very sparse. <br /><br />
Thus, instead of using bag-of-words, I choose to use the number of times a word occurs in a comment as a feature. The approach I choose to implement is the TF-IDF model. <br /><br />
Term Frequency (TF) = (Number of times term t appears in a document)/(Number of terms in the document)<br /><br />
Inverse Document Frequency (IDF) = log(N/n), where, N is the number of documents and n is the number of documents a term t has appeared in. The IDF of a rare word is high, whereas the IDF of a frequent word is likely to be low. Thus having the effect of highlighting words that are distinct. <br /><br />
We calculate TF-IDF value of a term as = TF * IDF <br /><br />
TFIDF gives more weightage to the words that are distinct. 

TF-IDF contains information on the more important words and the less important ones as well. So all data points are presented in a more meaningful way. <br />
Python scikit-learn library provides functions to calculate TF-IDF of text vocabulary given a text corpus.<br /><br /><br />
